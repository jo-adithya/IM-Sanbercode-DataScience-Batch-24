```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
titles = ['tanX', 'sinX', 'cosX']
fig, ax = plt.subplots(figsize=(12,5), nrows=1, ncols=3, sharex=True, sharey=True)
ax[0].plot(x, tan, 'b')
ax[2].plot(x, cos, 'g')
ax[1].plot(x, sin, 'r')
fig.tight_layout(pad=3.0)
fig.legend(titles, loc=9, bbox_to_anchor=(0.52, 1.14))

for i in range(3):
    ax[i].text(0.5, 14.5, 'Grafik ' + titles[i], fontsize=10, ha='center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
```

```python
##### code here
figure, axes = plt.subplots(figsize=(10, 6))
axes.plot(x, y)
axes.text(0, 1.565, 'Griafix CosX', ha='center')

axes_list = []
coor = [-0.1, 0.9, 0.9, -0.1]
colors = ['r', 'g', 'k', 'y']

for i in range(4):
    axes_list.append(axes.inset_axes([coor[i], coor[-i], 0.3, 0.3]))
    axes_list[-1].plot(x, y, 'b', linewidth=1)
    axes_list[-1].plot(x, np.cos(x ** (i + 2)), colors[i], linewidth=1)
    axes_list[-1].text(0, 1.2, f'Griafix CosX^{i + 2}', ha='center')
```
